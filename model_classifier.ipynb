```python
import pandas as pd
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

#train_data_file = './train_data02.csv'
#train_data_file = './feature_importance/train_data_50.csv'
#train_data_file = './feature_importance/train_twoStep_50.csv'
train_data_file = 'D:/QQFILE/Server_User_model/dataset/trainFinal4_.csv'
# test_data_file = './feature_importance/Final_val.csv'
train_data = pd.read_csv(train_data_file, encoding='utf-8')
# test_data = pd.read_csv(test_data_file, encoding='utf-8')
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', 10)
print(train_data.shape)
#print(test_data.shape)
train_data.head()
```

```python
features_columns = [col for col in train_data.columns if col not in ['pax_name', 'pax_passport', 'emd_lable', 'emd_lable2']]
#features_columns = [col for col in train_data.columns if col not in [ 'emd_lable2']]
#features_columns = [col for col in train_data.columns if col not in [ 'emd_lable','emd_lable2']]
#features_columns_test = [col for col in test_data.columns ]

train = train_data[features_columns].values
target =train_data['emd_lable2'].values

# test = test_data[features_columns].values
#target =train_data['emd_lable'].values
# target_test = test_data['emd_lable2'].values
```

```python
for i in range(train.shape[0]):
    for j in range(train.shape[1]):
        if train[i][j]<0:
            train[i][j]=0
```

```python
from sklearn.feature_selection import SelectKBest
from array import array
from sklearn.feature_selection import chi2

#skb = SelectKBest(chi2, k=70)
train = SelectKBest(chi2, k=90).fit_transform(train, target)
#train = skb.fit_transform(train, target)
#print(skb.get_support(indices=True))
#train.get_support(indices=True).tolist() 
print(train.shape)

train = pd.DataFrame(train)
# train.head(25)
#print(train.columns.values)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

clf = RandomForestClassifier(n_estimators=500, 
                             max_depth=70, 
                             random_state=0, 
                             n_jobs=-1)

X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.3, random_state=0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
#print(test.shape, target_test.shape)

#
clf = clf.fit(X_train, y_train)

print('=======')
y_true, y_pred = y_test, clf.predict(X_test)
#print(y_pred)
print(classification_report(y_true, y_pred))
print("RandomForest:", clf.score(X_test, y_test))
print('=======')

#print(clf.predict(X_test))

#
threshold = 0.39
predicted_proba = clf.predict_proba(X_test)
print(predicted_proba.shape)
predicted = (predicted_proba [:,1] >= threshold).astype('int')#
accuracy = accuracy_score(y_test, predicted)
print(classification_report(y_test, predicted))
print("accuracy: ", accuracy)

pred_1d = predicted_proba[:,1].reshape(X_test.shape[0],)
y_test = y_test.reshape(X_test.shape[0],)
print(pred_1d.shape, y_test.shape)
print("AUC: ",roc_auc_score(y_true, pred_1d))


print(predicted.shape)
#predicted.reshape(9373,2)
# data1 = predicted_proba.reshape(7030,2)

#print(train.shape, target.shape)
#print(test.shape, target_test.shape)

#
# print("test:")
# clf.score(test, target_test)

# print(clf.predict(test))
# #data1 = clf.predict(test)
# #print(clf.predict_proba(test))
# data2 = clf.predict_proba(test)
# data2.reshape(6771,2)
```
