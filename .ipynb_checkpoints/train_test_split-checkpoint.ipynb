```python
import pandas as pd
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

#train_data_file = './train_data02.csv'
#train_data_file = './feature_importance/train_data_50.csv'
#train_data_file = './feature_importance/train_twoStep_50.csv'
train_data_file = './feature_importance/Final_train.csv'
#train_data_file = './feature_importance/trainFinal4.csv'
#train_data_file = './feature_importance/Final_train_downsample.csv'
test_data_file = './feature_importance/Final_val.csv'
train_data = pd.read_csv(train_data_file, encoding='utf-8')
test_data = pd.read_csv(test_data_file, encoding='utf-8')
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', 10)
print(train_data.shape)
print(test_data.shape)

train_data.head()
```

```python
features_columns = [col for col in train_data.columns]
#features_columns = [col for col in train_data.columns if col not in ['pax_name', 'pax_passport', 'emd_lable2']]
#features_columns_test = [col for col in test_data.columns ]
#features_columns = [col for col in train_data.columns if col not in ['pax_name', 'pax_passport', 'emd_lable2', 'label3']]

train = train_data[features_columns].values
target =train_data['emd_lable2'].values
#target =train_data['label3'].values
```

```python
from sklearn.model_selection import train_test_split
import pandas
from pandas import DataFrame

X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.27, random_state=0)

y_train = y_train.reshape((-1,1))
y_test = y_test.reshape((-1,1))

train = np.hstack((X_train,y_train))
test = np.hstack((X_test,y_test))
train_csv = DataFrame(train)
test_csv = DataFrame(test)

train_csv.to_csv('./feature_importance/train_stacking_train.csv',index=False)
test_csv.to_csv('./feature_importance/train_stacking_test.csv',index=False)
```
