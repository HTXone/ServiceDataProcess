```python
def find_repeat_data(name_list):
    """
    查找列表中重复的数据
    :param name_list: 
    :return: 一个重复数据的列表，列表中字典的key 是重复的数据，value 是重复的次数
    """
    repeat_list = []
    for i in set(name_list):
        ret=name_list.count(i) # 查找该数据在原列表中的个数
        if ret > 1:
            item=dict()
            item[i] = ret
            repeat_list.append(item)
    print(repeat_list)
    return repeat_list
find_repeat_data(list1)
```
